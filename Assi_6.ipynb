```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
```

```python
# reading the database
data = pd.read_csv("udemy-courses-web-development.csv")
```

```python
# printing the top 10 rows
display(data.head(10))
```

```python
data.isnull().sum()
```

```python
data = data.fillna(0) #filling missing values with 0
```

```python
display(data.head(10))
```

```python
data.isnull().sum()
```

```python
# Scatter plot with level against num_reviews
sns.scatterplot(x='level', y='num_reviews', data=data,
               hue='Price_model')
plt.show()
```

```python
# Scatter plot with day against tip
plt.scatter(data['level'], data['num_reviews'], c=data['rating'],
            s=data['num_lectures'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('level')
plt.ylabel('num_reviews')
 
plt.colorbar()
 
plt.show()
```

```python
# Scatter plot with day against tip
plt.plot(data['num_lectures'])
plt.plot(data['content_duration'])
```

```python
# Bar chart with level against num_reviews
plt.bar(data['level'], data['num_reviews'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('level')
plt.ylabel('num_reviews')
```

```python
# histogram of rating
plt.hist(data['rating'])
plt.title("Histogram") 
```

```python
# draw lineplot
sns.lineplot(x="Price_model", y="rating", data=data)
 
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
```
